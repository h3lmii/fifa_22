```python
def class_age(age):
  if age<21:
    return 1
  elif age<30:
    return 3
  else:
    return 0.5

class_age(37)
```

```python
def class_wage(wage):
  if wage<100000:
    return 5
  elif wage<200000:
    return 4
  elif wage<300000:
    return 3
  else:
    return 1

class_wage(370000)
```

```python
df1['ClassAge']=df1.Age.apply(class_age)
df1['ClassAge']
```

```python
df1.Wage
```

```python
df1.Value
```

```python
def KM(s):
  s=s.replace('€', '')
  if s.endswith('M'):
    s=s.replace('M','')
    s=float(s)*1000000
  elif s.endswith('K'):
    s=s=s.replace('K','')
    s=float(s)*1000
  else:
    s=float(s)
  return s
KM("€925M")
```

```python
df1.Wage=df1.Wage.apply(KM)
df1.Wage
```

```python
df1.Value=df1.Value.apply(KM)
df1.Value
```

```python
def class_value(v):
  v=v/100000
  if v<1:
    return 1
  elif v<30:
    return 3
  else:
    return 0.5

class_value(37)
```
